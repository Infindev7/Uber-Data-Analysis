Importing required python modules

```python
import numpy as np
import pandas as pd
```

Storing data from file in variable 'df'

```python
df=pd.read_csv('uber-raw-data-apr14.csv')
```

```python
type(pd)
```

Checking first few rows of dataframe

```python
df.head()
```

Checking missing values

```python
df.isna().sum()
```

Removing all missing values

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

Convertng a column of string dates into actual datetime objects

```python
df['Date/Time'] = pd.to_datetime(df['Date/Time'])
```

Extracting specific components (Month, Year, Day, and Time) from the datetime column

```python
df['Month'] = df['Date/Time'].dt.month
df['Year'] = df['Date/Time'].dt.year
df['Day'] = df['Date/Time'].dt.day
df['Time'] = df['Date/Time'].dt.time
```

Again checking first few rows of dataframe

```python
df.head()
```

Dropping 'Date/Time' column

```python
df = df.drop('Date/Time', axis=1)
```

Arranging the columns

```python
df = df[['Year','Month','Day','Time','Lat','Lon','Base']]
```

Viewing final dataframe

```python
df.head()
```

Saving the dataFrame into a new CSV file

```python
df.to_csv('April_14.csv')
```

Convertng a column of string dates into actual datetime objects and extracting specific components (Month, Year, Day, and Time) from the datetime column

```python
df['Date/Time'] = pd.to_datetime(df['Date/Time'])
df['Month'] = df['Date/Time'].dt.month
df['Year'] = df['Date/Time'].dt.year
df['Day'] = df['Date/Time'].dt.day
df['Time'] = df['Date/Time'].dt.time
```

```python
df.head()
```

Dropping 'Date/Time' and 'locationID' columns

```python
df = df.drop('Date/Time', axis=1)
df = df.drop('locationID', axis=1)
```

```python
df.head()
```

Arranging the columns and viewing the final dataframe

```python
df = df[['Year','Month','Day','Time','Lat','Lon','Dispatching_base_num','Affiliated_base_num']]
df.head()
```

Saving the dataFrame into a new CSV file
